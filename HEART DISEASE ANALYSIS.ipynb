# SPLITTING TRAIN AND TEST DATASET

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    data.drop(labels=['target'], axis=1),
    data['target'],
    test_size=0.3,
    random_state=44)

X_train.shape, X_test.shape

##splitting the dataset into train and test dataset where test contains 0.3% of total dataset
```

```python
X_train.drop(['age', 'sex','trestbps','chol'],axis=1,inplace=True)
#dropping columns from X_train
```

```python
X_test.drop(['age', 'sex','trestbps','chol'],axis=1,inplace=True)
#dropping columns from X+_test
```

```python
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
corr_features = correlation(X_train, 0.8)
(corr_features)
```

```python
corr_features = correlation(X_test, 0.8)
(corr_features)
```

```python
X_train.drop(labels=corr_features, axis=1,inplace=True)
X_test.drop(labels=corr_features, axis=1,inplace=True)

X_train.shape,X_test.shape

##new shape train and test after dropping columns
```

### IMPORTING LIBRARIES FOR MODELLING

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
from sklearn import svm 
from sklearn.svm import SVC
```

### LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
predictions = logmodel.predict(X_test)
print("Accuracy Score:",logmodel.score(X_test, y_test))
print("Confusion Matrix:",confusion_matrix(y_test, predictions))
print("Classification Report:",classification_report(y_test,predictions))
```

### STATIC VIRTUAL MACHINE

```python
def svc_param_selection(X, y, nfolds):
    Cs = [0.001, 0.01, 0.1, 1, 10]
    gammas = [0.001, 0.01, 0.1, 1]
    param_grid = {'C': Cs, 'gamma' : gammas}
    grid_search = GridSearchCV(svm.SVC(kernel='rbf'), param_grid, cv=nfolds)
    grid_search.fit(X, y)
    grid_search.best_params_
    return grid_search.best_params_
```

```python
svc_param_selection(X_train,y_train,4)
```

```python
C=1.0
svc=SVC(kernel='rbf',C=10,gamma=0.001)
svc.fit(X_train,y_train)
predictions2=svc.predict(X_test)
print("Accuracy Score:",svc.score(X_test, y_test))
print("Confusion Matrix:",confusion_matrix(y_test, predictions2))
print("Classification Report:",classification_report(y_test,predictions2))
```

### RANDOM FOREST
